```
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/stock_data/stocks_yahoo.csv')

df.head()
```

```
#import fix_yahoo_finance as yf
#dg = yf.download(tickers=["AAPL","AMZN","JPM","GOOG","ORCL"], start="2018-01-01", end="2019-11-20", group_by="ticker")
#dg.head(10)
co = pd.read_csv('https://raw.githubusercontent.com/datasets/s-and-p-500-companies/master/data/constituents.csv')
co.tail()
```

```
co.isnull().sum()
```

```
## prep both datasets for combining and get rid of potential duplicate company name column in co dataset
co.drop(columns=['Name'], inplace=True)
data = df['company_name'].str.split("- ", n = 1, expand = True)
df['Symbol'] = data[0]
df['company_name'] = data[1]
df.drop(columns=["no"], inplace=True)
df['date'] = pd.to_datetime(df['date'], format='%b %d %Y')
df.head()
```

```
df.isnull().sum()
```

```
stock_merge = pd.merge_ordered(df, co, fill_method='ffill', on='Symbol')
stock_merge
```

```
##drop empty rows or rows with NaN
stock_row_drop = stock_merge.dropna(how='any')
stock_row_drop.isnull().sum()
```
